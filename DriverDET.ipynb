```python
import torch 
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
torch.cuda.is_available()
```

```python
model=torch.hub.load('ultralytics/yolov5','yolov5s')
```

```python
cap=cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame=cap.read()
    results=model(frame)
    cv2.imshow('Driver State Detection System', np.squeeze(results.render()))

    if cv2.waitKey(10) & 0xFF == ord('f'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import uuid
import os
import time
```

```python
IMAGES_PATH= os.path.join('yolov5','data','images')
labels = ['awake', 'drowsy','head bobbing right','head bobbing left','yawning','distracted right','distracted left']
number_imgs=100
```

```python
cap = cv2.VideoCapture(0)

for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(10)
    
    
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        
    
        ret, frame = cap.read()
        
        
        imgname=os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg')
        
         
        cv2.imwrite(imgname,frame)
        
        
        cv2.imshow('Image Collection', frame)
        
        time.sleep(1)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
